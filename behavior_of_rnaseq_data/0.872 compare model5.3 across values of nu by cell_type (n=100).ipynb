## Summarize fit across different values of `nu`

```python
loo_results = {1: loo_nu1, 3: loo_nu3, 4: loo_nu4, 5: loo_nu5}
```

```python
loo_nu1.pointwise.head()
```

```python
loodf = list()
for (nu, loores) in loo_results.items():
    thisdf = loores.pointwise.reset_index()
    thisdf['nu'] = nu
    loodf.append(thisdf)
loodf = pd.concat(loodf)
```

```python
loodf.head()
```

```python
loodf.sort_values(['index','nu'], inplace=True)
loodf['first_diff'] = loodf.groupby('index')['pointwise_elpd'].diff()
loodf.fillna(0, inplace=True)
loodf.head()
```

```python
loodfw = loodf.pivot(columns='nu', values='pointwise_elpd', index='index')
```

```python
loodfw.head()
```

```python
sns.boxplot(data=loodf, x='nu', y='first_diff', fliersize=0)
```

```python
from matplotlib import pyplot as plt
for i in np.arange(max(loodf.index)):
    plt.plot(loodf.loc[loodf['index'] == i, 'nu'],
             loodf.loc[loodf['index'] == i, 'first_diff'],
             alpha=0.2)
```

```python
sample_df.head()
```

```python
loodf2 = pd.merge(sample_df, loodf, on='index')
```

```python
sns.boxplot(data=loodf2, x='nu', y='first_diff', hue='SubSet', fliersize=0)
_ = plt.ylim(-0.2,0.2)
```

```python
sns.boxplot(data=loodf2, x='nu', y='first_diff', hue='sample_id', fliersize=0)
_ = plt.ylim(-0.2,0.2)
_ = plt.legend([])
```
